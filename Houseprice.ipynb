```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
X = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]).reshape((-1,1))
Y= np.array([20728, 27329, 30859, 28422, 26703, 24931, 27445, 29156, 31391, 39108, 50151, 48099, 55017, 63203, 74530 ])
```

```python

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=.25,random_state=10)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
print(f'Records in training data: {x_train.shape[0]:,}')
print(f'Records in test data: {x_test.shape[0]:,}')
print('\nFeatures:')
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
print('intercept:', lm.intercept_)
print('slope:', lm.coef_)
y_pred=lm.predict(x_test)
#print('predicted response:', y_pred)
```

```python
yhat_train = lm.predict(x_train)
print(yhat_train[0:5])
yhat_test = lm.predict(x_test)
print(yhat_test[0:5])
```

```python
sns.distplot(Y,kde = True, color ='red')
```

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
  width = 12
  height = 10
  plt.figure(figsize=(width, height))
  ax1 = sns.distplot(RedFunction, hist=False, color="r", label=RedName)
  ax2 = sns.distplot(BlueFunction, hist=False, color="b", label=BlueName, ax=ax1)
  plt.title(Title)
  plt.xlabel('Goldprice')
  plt.ylabel('Year')
  plt.legend()
  plt.show()
  plt.close()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
Title = 'Distribution Plot of Predicted Value Using Training Data vs Training Data Distribution'
DistributionPlot(y_train, yhat_train, "Actual Values (Train)", "Predicted Values (Train)", Title)
Title='Distribution Plot of Predicted Value Using Test Data vs Data Distribution of Test Data'
DistributionPlot(y_test,yhat_test,"Actual Values (Test)","Predicted Values (Test)",Title)
```

```python
#We use scikit learn to import the linear regression model.
#we fit the model on the training data and predict the values for the testing data. We use R2 score to measure the accuracy of our model.

print('R2 Score:',r2_score(y_test,y_pred))
#r_sq = lm.score(y_test,y_pred)
#print('coefficient of determination:', r_sq)
plt.scatter(y_test,y_pred)
plt.xlabel("Prises:$y-i$")
plt.ylabel("Predicted Prises:$\hat{y}-i$")
plt.title("Prises vs Predicted Prises:$y-i$ vs $\hat{y}-i$")
plt.show()
#R2 Score: 0.7063361703359012
```
